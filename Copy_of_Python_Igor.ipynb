# üêç Parte 1: Fundamentos e Tipagem Din√¢mica

Em Python, vari√°veis s√£o "r√≥tulos" que colamos em valores na mem√≥ria. Diferente de linguagens como Java ou C, o Python √© **dinamicamente tipado**. Isso significa que voc√™ n√£o precisa declarar o tipo (int, string) antes; o Python infere isso no momento da atribui√ß√£o.

### üîπ Regras de Ouro
* **Nomes de vari√°veis:** Use `snake_case` (tudo min√∫sculo, separado por underline). Ex: `nome_usuario`, `total_vendas`.
* **Case Sensitive:** `Idade` √© diferente de `idade`.
* **Atribui√ß√£o:** O sinal `=` serve para guardar um valor.

### üîπ Os Tipos Primitivos
1. **`str` (String):** Texto. Sempre entre aspas simples `' '` ou duplas `" "`.
2. **`int` (Inteiro):** N√∫meros sem casa decimal (positivos ou negativos).
3. **`float` (Ponto Flutuante):** N√∫meros com casa decimal. **Aten√ß√£o:** Use **ponto** `.` e n√£o v√≠rgula.
4. **`bool` (Booleano):** L√≥gica bin√°ria. Apenas `True` (Verdadeiro) ou `False` (Falso). **Aten√ß√£o:** A primeira letra √© sempre mai√∫scula!

```python
# Declarando vari√°veis
curso = "Python Backend"     # String
ano = 2025                   # Int
preco_mensalidade = 150.50   # Float
inscricoes_abertas = True    # Bool

# A fun√ß√£o type() mostra qual o tipo do dado
print(type(curso))
print(type(preco_mensalidade))

# Podemos mudar o tipo de uma vari√°vel a qualquer momento (Tipagem Din√¢mica)
ano = "Dois mil e vinte e cinco"
print(ano)
```

# 2. Entrada, Sa√≠da e F-Strings

### üó£Ô∏è Interagindo com o usu√°rio (`input`)
A fun√ß√£o `input()` pausa o programa e espera o usu√°rio digitar algo.

> **‚ö†Ô∏è Cuidado:** O `input` sempre retorna o dado como **TEXTO (String)**. Se voc√™ pedir a idade e o usu√°rio digitar `20`, o Python entender√° como `"20"`. Para fazer contas, voc√™ precisa converter (*fazer o casting*) usando `int()` ou `float()`.

### üé® Formatando Textos (`f-strings`)
A forma mais moderna e profissional de misturar texto com vari√°veis √© usando o `f` antes das aspas. Isso permite colocar vari√°veis direto dentro das chaves `{}` (semelhante ao `${}` do JavaScript).

**Exemplo:**
`print(f"O aluno {nome} tirou nota {nota}")`

```python
# Exemplo de Entrada e Convers√£o
nome = input("Digite seu nome: ")
idade_texto = input("Digite sua idade: ")

# Convertendo de String para Inteiro
idade = int(idade_texto)

# C√°lculo simples
ano_nascimento = 2025 - idade

# Sa√≠da formatada com f-string
print(f"Ol√°, {nome}! Se estamos em 2025, voc√™ provavelmente nasceu em {ano_nascimento}.")
```

# 3. Listas (Arrays)

Listas s√£o estruturas ordenadas e mut√°veis. √â a ferramenta mais vers√°til do Python.

* **√çndice:** Come√ßa sempre em `0`.
* **√çndice Negativo:** O Python permite acessar de tr√°s para frente. `-1` √© o √∫ltimo item, `-2` o pen√∫ltimo.
* **Fatiamento (Slicing):** Podemos pegar um peda√ßo da lista usando `lista[inicio:fim]`. **Nota:** O √≠ndice final **n√£o** √© inclu√≠do.

### M√©todos Importantes:
* `lista.append(item)`: Adiciona ao final (como o *push*).
* `lista.pop()`: Remove o √∫ltimo item.
* `lista.remove(valor)`: Remove a primeira ocorr√™ncia de um valor espec√≠fico.


```python
tecnologias = ["HTML", "CSS", "JavaScript", "Python"]

# Acessando
print(f"A primeira linguagem √©: {tecnologias[0]}")
print(f"A √∫ltima linguagem √©: {tecnologias[-1]}")

# Modificando
tecnologias.append("SQL")  # Adiciona no fim
tecnologias[1] = "Tailwind" # Substitui o CSS

# Slicing (Pegando do √≠ndice 1 at√© o 3 - o 3 n√£o entra)
backend = tecnologias[2:4] # Pega JS e Python

print(f"Lista atualizada: {tecnologias}")
print(f"Sub-lista: {backend}")
```

# 4. Dicion√°rios (Key-Value)

Se as Listas s√£o como arrays, os **Dicion√°rios** s√£o equivalentes aos Objetos Literais do JavaScript ou JSON. Eles servem para estruturar dados complexos.

Cada item possui uma **chave** (√∫nica) e um **valor**.

* **Sintaxe:** Usa chaves `{}`. Ex: `{"chave": "valor"}`.
* **Acesso:** Pela chave. Ex: `usuario["nome"]`.
* **Diferen√ßa pro JS:** As chaves **precisam** estar entre aspas no Python.

```python
produto = {
    "id": 450,
    "nome": "Notebook Gamer",
    "preco": 4500.00,
    "em_estoque": True
}

# Acessando valores
print(f"O produto {produto['nome']} custa R$ {produto['preco']}")

# Adicionando nova chave
produto["marca"] = "Dell"

# Editando valor
produto["preco"] = 4200.00

print(produto)
```

# 5. Estruturas de Controle (Fluxo)

No Python, **a indenta√ß√£o (espa√ßamento) √© obrigat√≥ria**.
N√£o usamos chaves `{}` para definir blocos de c√≥digo (como no `if` ou `for`), usamos os **dois pontos `:`** seguidos de c√≥digo indentado (geralmente 4 espa√ßos ou 1 Tab).

### üîÄ Condicionais (`if`, `elif`, `else`)
Verifica condi√ß√µes l√≥gicas. O `else if` do JavaScript aqui vira `elif`.

### üîÑ Loops (`for`)
O `for` do Python √© um "for-each". Ele percorre cada item de uma cole√ß√£o (seja uma lista, uma string ou um range de n√∫meros). √â mais simples que o `for(i=0; i<10; i++)` cl√°ssico.

```python
notas = [5.5, 8.0, 9.5, 4.0]

print("--- Boletim ---")

for nota in notas:
    if nota >= 7:
        print(f"Nota {nota}: APROVADO")
    elif nota >= 5:
        print(f"Nota {nota}: RECUPERA√á√ÉO")
    else:
        print(f"Nota {nota}: REPROVADO")
```
